```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread("./Group_photo.jpg")
```

```python
plt.imshow(img)
```

```python
##haarCascade file   -> xml file
## cb.lk/facemodel
## help detect faces
## download the file from this link
## you can google haarCascade fils and get different detection codes for differnt things on github
```

```python
detector = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
faces=detector.detectMultiScale(img)
```

```python
faces
```

```python
faces.shape
```

```python
##means it has detected 7 faces which each has 4 properties
```

```python
##209 is x coordinate 40 is ycoordinate of startimg point and 26 is width and last 26 is height
```

```python
## plotting rectangle around detected face 1
```

```python
img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,255),2)
img
```

```python
plt.imshow(img)
```

```python
#plotting rectangle for all faces
```

```python
for face in faces:
  x,y,w,h = face
  img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,255),2)
```

```python
plt.imshow(img)
```
